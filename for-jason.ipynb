# Task 1.1 - Data Preparation

```python
import pandas as pd

import numpy as np

import sklearn as sk
```

```python
#Â Reading the csv file

read_test = pd.read_csv('glass_dataset.csv')

targets_type = read_test.values[:, -1]

inputs = read_test.values[:, :-1]
```

```python
# Printing out the inputs and targets(type) data file

print(targets_type)

print(inputs)
```

```python
# Find the smallest and largest value

# Find out the range

# For each element divide the element by the range and minus the minumum

def normalisation(inputs):

    colums_count = inputs.shape[1]

    for i in range(colums_count):

        minimum_value = np.min(inputs[:, i])

        maximum_value = np.max(inputs[:, i])

        range_value = maximum_value - minimum_value
        #Finding out the range

        inputs[:,i] = (inputs[:,i] - minimum_value) / range_value

    return inputs



inputs = normalisation(inputs)

print(inputs[:5])
```

# Task 1.2 - Classification

```python
#import important libraries

import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
 
from sklearn.neighbors import KNeighborsClassifier #knn library

from sklearn.svm import SVC # svm library

from sklearn import datasets

```

```python
# Train the knn classifier .

def knn_train(k):

    classifier = KNeighborsClassifier(n_neighbors = k)
    
    classifier.fit(inputs,targets_type)

    classifiedData = classifier.predict(inputs)

    return classifiedData

knn_train(20)
```

```python
#Function that adds 20 times to array knn

def array_knn_trains():

    scores = []

    for k in range(1,20):

        knn = KNeighborsClassifier(n_neighbors = k)

        knn.fit(inputs,targets_type)

        scores.append(knn.score(inputs,targets_type))
    
    return scores

scores = array_knn_trains()




```

```python
#Plotting knn data on the graph

plt.figure()

plt.xlabel('K')

plt.ylabel('Accuracy')

plt.plot(scores)



plt.savefig("knnloop.png", bbox_inches = "tight")
```

```python

# Function

def array_svm_trains():

    array = []

    

    new_svc = SVC(kernel="linear")
    SVC(kernal="radial")
    SVC(kernal="polynomial", degree=2)
    SVC(kernal="polynomial", degree=3)

    new_svc.fit(inputs, targets_type)

    array.append(new_svc.score(inputs,targets_type))
    
    return scores

array = array_knn_trains()






```

```python
#Fitness function Number 1 - Schaffer Function

def schaffer(x, y):

    numerator = np.sin(x**2 - y**2)**2 - 0.5

    denominator = (1 + 0.001 * (x**2 + y**2))**2
    
    s_function = 0.5 + (numerator / denominator)

    return s_function
```

```python
#Fitness function Number 2 - Modified Ackley Function

def modified_ackley(x, y):

    m_a_function = -20 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2))) - np.exp(0.5 * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y))) + 20
    
    return m_a_function
```

```python
#Fitness function Number 3 - Easom Function

def easom(x, y):

    e_function = -np.cos(x) * np.cos(y) * np.exp(-((x - np.pi)**2 + (y - np.pi)**2))

    return e_function
```
